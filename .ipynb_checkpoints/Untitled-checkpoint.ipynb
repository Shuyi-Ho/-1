```python
import pandas as pd 
```

```python
df_gdp= pd.read_table("gdp.csv",   
                       encoding = 'utf8',      
                       header = 3,       
                       skiprows = [35,36],    
                       sep =','             
                      )
df_edu= pd.read_table("education.csv",  
                       encoding = 'utf8',      
                       header = 3,             
                       skiprows = [35,36],     
                       sep =','                 
                      )
```

```python
def prepare(df_in, l_variable, l_value):
    df = df_in.copy()
    df.columns = [ int(x.replace('年','')) if x!='地区' else x for x in df.columns]
    years = [ x for x in df.columns if x!='地区'] 
    df_out = pd.melt(df, id_vars=['地区'], value_vars=years)
    df_out.columns = ['地区', l_variable, l_value]
    df_out = df_out.set_index(['地区',  l_variable])
    return (df_out)
```

```python
prepare(df_gdp, "年", "gdp")
```

```python
prepare(df_edu, "年", "高校数量")
```

```python
df_ggg = prepare(df_gdp, "年", "gdp").join( prepare(df_edu, "年", "高校数量"), 
                                          how = "left" )
df_ggg
```

```python
%matplotlib inline
df_ggg.query("年==2015").plot(kind='scatter', x='高校数量', y='gdp', 
          title = '高校数量与gdp的关系：分省数据')
```

```python
import numpy as np
x = df_ggg.query("年==2015")['高校数量']
y = df_ggg.query("年==2015")['gdp']
fit = np.polyfit(x=x , y=y, deg=1)

ax = df_ggg.query("年==2015").plot(kind='scatter', x='高校数量', y='gdp', 
          title = '高校数量与gdp的关系：分省数据')
ax.plot(x, fit[0] * x + fit[1], color='red')
ax.scatter(x, y)
```

```python
df_ggg['desc'] = df_ggg.index.get_level_values(0)

from bokeh.plotting import figure, output_file, show, output_notebook
from bokeh.models import HoverTool, BoxSelectTool

title = '分省数据' 
xlabel = 'edu：高校数量' 
ylabel = 'gdp：产值'

hover = HoverTool(
    tooltips=[
        ( '高校数量',  '$x{0,F}'), 
        ( '产值',  '$y{0,F}'),
        ( '地区',  '@desc'  ),
    ]
)

TOOLS = [hover, "pan, wheel_zoom, box_zoom, reset, resize"]
fig = figure(plot_width=800, plot_height=600,
           x_axis_label = xlabel, y_axis_label = ylabel, 
           title=title, tools=TOOLS)

df_plot = df_ggg.query("年==2015")

fig.circle("高校数量", "gdp",
         source = df_plot,
         fill_alpha=0.6, line_color=None, size= 15)

xs = [min(df_plot["高校数量"]), max(df_plot["高校数量"])]
fig.line(x=xs, y=[ fit[0] * x + fit[1] for x in xs] )

output_notebook() 

show(fig)
```
